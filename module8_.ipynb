Q2 - Explain the difference between Absolute and Relative positioning.

```python
1. Relative Positioning (`position: relative`):

   - Moves the element relative to its original position in the normal document flow.
   - The element still occupies space in the layout as if it were not moved.
   - Useful for making small adjustments or as a reference container for absolutely positioned children.

2. Absolute Positioning (`position: absolute`):

   -  Removes the element from the normal document flow entirely.
   -  Positions the element relative to the nearest positioned ancestor (i.e., one with `position: relative`, `absolute`, or `fixed`); if none, it uses the `<html>` element.
   -  Does not leave a space where it was originally placed.

3. Stacking and Layering:
  - Both can be layered using `z-index`, but since `absolute` is removed from flow, it often stacks on top of `relative`.

4. **Child-Parent Behavior**:
   -  An `absolute` element placed inside a `relative` element will position itself relative to that parent, which is commonly used in layouts.


```

Q5 - Explain the z-index, with a code example.


```python
""" z-index is a CSS property that controls the stacking order of elements along the z-axis (the axis going in and out of the screen). It only works on positioned elements (position: relative, absolute, fixed, or sticky).

Higher z-index values appear on top of lower ones.

The default value is auto (which is treated as 0 if compared).

z-index only works if the element has a position other than static. """



<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <title>z-index Example</title>
  <style>
    .box {
      width: 200px;
      height: 200px;
      position: absolute;
      top: 50px;
      left: 50px;
    }

    .box1 {
      background-color: red;
      z-index: 1;
    }

    .box2 {
      background-color: green;
      top: 100px;
      left: 100px;
      z-index: 2;
    }

    .box3 {
      background-color: blue;
      top: 150px;
      left: 150px;
      z-index: 0;
    }
  </style>
</head>
<body>

  <div class="box box1"></div>
  <div class="box box2"></div>
  <div class="box box3"></div>

</body>
</html>
```
